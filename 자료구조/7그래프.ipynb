# 그래프

- 그래프는 정점(vertex)과 간선(edge)으로 이루어진 자료구조이다.
- 그래프에서 자료를 담고 있는 원소를 모두 정점(V) 혹은 노드라고 한다.
- 간선은 정점 두 개를 연결하는 연결선의 역할을 한다. ( 간선에는 가중치 개념O(비용, 거리라고 생각하면 될 것 같다) - 양수간선, 음수간선)
- 차수(degree)는 하나의 정점과 연결되어 있는 간선의 개수이다.
- 사이클(cycle) : 같은 정점을 두 번 이상 방문한 경우 본 그래프에 '사이클'이 존재한다고 한다. (원이 그려지는..우리가 아는 그 사이클이라고 생각하면 쉽다!) > 사이클이 존재하지 않는 그래프는 `트리(tree)`라고 한다.
- 그래프는 여러 알고리즘에 많이 사용되는 자료구조! 코테에 많이 사용된다!

## 1. 유니온 파인드(union-find)
- 그래프의 사이클이 생성되는지 판별하는 알고리즘
- 유니온 파인드 = union연산 + find연산
- union연산으로 두 정점, 두 노드를 연결해준다. >> 항상 루트노드끼리 연결해준다.
- find연산으로 루트노드, 대표노드를 찾아서 두 노드가 한 그래프 안에 있는지, 두 노드간의 경로가 존재하는지를 판단한다.
- 유니온 파인드 더 자세한 내용: https://chiefcoder.tistory.com/55 참고하기!

## 2. 위상 정렬(Topological-Sorting)
- 선조건: 사이클 없고, 방향성이 있는 그래프일 때 위상정렬 알고리즘을 사용할 수 있다.(사이클x, 방향o)
- 위상 정렬: 그래프의 각 노드들을 `선형으로 정렬`!!
- 위상 정렬 했을 때 만들어지는 그래프는 유일하지 않다.(답이 하나가 아니다.)
- 위상 정렬 대표 문제: 수강신청(전후관계o ex. 선수과목)
- 위상정렬 더 자세한 내용: https://velog.io/@kimdukbae/%EC%9C%84%EC%83%81-%EC%A0%95%EB%A0%AC-Topological-Sorting 참고하기!

## 3. 다익스트라 / 벨만-포드 / 플로이드-워셜
- 세 알고리즘의 공통점: `최단거리를 구하는 알고리즘`
1. 다익스트라
- 시작점이 있고, 시작점에서 다른 모든 노드로 가는 최단거리를 구하는 알고리즘
- 단, 음수간선(간선의 가중치가 음수인 것)이 있으면 이 알고리즘 사용x

2. 벨만-포드
- 시작점이 있고, 시작점에서 다른 모드 노드로 가는 최단거리를 구하는 알고리즘
- 다익스트라와 거의 같다. 다른점은 단 하나! >> `음수간선이 있어도 사용O`
- 음수간선이 있다면 벨만-포드를 사용하면 된다.

3. 플로이드-워셜
- `시작점이 없고`, 모든 노드에 대한 최단거리를 구하는 알고리즘
- 시작점이 없기 때문에 모든 노드에 대해서 탐색해야 해서 시간 복잡도가 좋지 않다.
- 시작점이 없고, 탐색하는 노드 개수가 낮을때 플루이드-워셜 알고리즘 사용! > ex. 모든 도시 간의 최단거리 구해라(단, 도시의 개수는 100개 이하)

## 4. 최소 신장 트리(Minimum Spanning Tree, MST)
- 모든 노드를 연결하되, 연결할 때 간선의 가중치가 최소가 되게끔 구하는 것이 최소 신장 트리 알고리즘이다. (최소의 비용을 써서 모든 노드를 연결하고 싶을 때 사용)
- 트리는 사이클x > 가중치를 최소로 하면 모든 노드를 연결하는 방법을 찾을 때, 사이클은 나오는 것을 방지하기 위해 유니온 파인드(사이클 여부 판단)를 함께 이용해 문제를 푼다.

## 총정리
- 유니온 파인드: `사이클 유무 판단`
- 위상 정렬
    - 선조건: 사이클x, 방향O>방향간선이어야 함 
    - `노드를 정렬`해주는 알고리즘
    - 정렬 결과가 꼭 1개는 아니다! 
    - 수강신청, 게임에서 빌드오더 등의 문제에서 사용
- 다익스트라, 벨만포드, 플로이드워셜: `최단거리 알고리즘`(노드에서 노드 사이의 최단거리를 구해주는 알고리즘)
    - 다익스트라: 시작점 있고(각 시작점에서 각 노드까지의 최단거리 구함), 음수간선x
    - 벨만포드: 시작점 있고, 음수간선o
    - 플로이드워셜: 시작점x, 임의의 모든 노드쌍에 대해 최단거리 구함
- 최소신장트리: `그래프에서 최소의 가중치 합으로 모든 노드를 연결할 수 있게 해주는 알고리즘` >> 사이클x(있을수가 없음!)(이 사이클 있으면 안되는거를 연산해주는 알고리즘이 유니온 파인드이기 때문에 최소신장트리 안에 유니온 파인드를 당연히 구현해 사용한다!) 

## 그래프의 표현 - 인접행렬, 인접리스트
1. 인접 행렬(Adjacency Matrix)
- 2차원 배열로 그래프의 연결 관계를 표현
- 노드 5개 > 5x5 인접행렬로 표현 (`5x5표` 라고 생각하면 쉬움)
- 시간도 오래걸리고, 공간도 많이 차지(시간복잡도&공간복잡도 bad)
- 하지만 구현하기 편하기 때문에 노드가 적을때(100,200,300개 등) 사용하면 good > 인접 행렬은 노드 개수에 따라 사용 여부를 적절하게 판단하는 능력 필요!

2. 인접 리스트(Adjancency List)
- 링크드 리스트로 그래프의 연결 관계를 표현
- 인접 리스트를 통해서 푸는 문제들이 제일 많다!(그만큼 많이 사용한다)
- 구현은 복잡하지만, 시간&공간복잡도 면에서 굉장히 효율적이다.(빠르고, 공간도 덜 차지)
